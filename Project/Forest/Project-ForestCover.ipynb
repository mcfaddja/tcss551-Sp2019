### Test/Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_features, test_features, train_types, test_types = train_test_split(features, types, test_size = 0.15, random_state = 42)
```

### Final Data Type Stuff

```python
dataCats = types.astype('category')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression().fit(train_features, train_types)
```

```python
linReg.score(test_features, test_types)
```

```python
linReg.score(train_features, train_types)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(train_types, linReg.predict(train_features))
```

```python
mean_squared_error(test_types, linReg.predict(test_features))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg1 = LogisticRegression(random_state=0, \
                            solver='lbfgs', \
                            multi_class='multinomial', \
                            max_iter = 10000, \
                            n_jobs = 6).fit(train_features, train_types)
```

```python
logReg2 = LogisticRegression(random_state=0, \
                            solver='lbfgs', \
                            multi_class='ovr', \
                            max_iter = 10000, \
                            n_jobs = 6).fit(train_features, train_types)
```

```python
logReg1.score(test_features, test_types)
```

```python
logReg1.score(train_features, train_types)
```

```python
logReg2.score(test_features, test_types)
```

```python
logReg2.score(train_features, train_types)
```

```python
mean_squared_error(train_types, logReg1.predict(train_features))
```

```python
mean_squared_error(test_types, logReg1.predict(test_features))
```

```python
mean_squared_error(train_types, logReg2.predict(train_features))
```

```python
mean_squared_error(test_types, logReg2.predict(test_features))
```
