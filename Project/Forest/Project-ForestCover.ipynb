### Final Data Type Stuff

```python
typeCats = types.astype('category')
```

### Test/Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_features, test_features, train_types, test_types = train_test_split(features, types, test_size = 0.15, random_state = 42)
```

```python
train_features, test_features, train_typeCats, test_typeCats = train_test_split(features, typeCats, test_size = 0.15, random_state = 42)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression().fit(train_features, train_typeCats)
```

```python
coeffs = linReg.coef_
indices1 = np.argsort(coeffs)[::-1]
```

```python
for f in range(features.shape[1]):
    print("%d. feature %d (%f) %s" % (f + 1, indices1[f], coeffs[indices1[f]], names[indices1[f]]))
```

```python
linReg.score(test_features, test_typeCats)
```

```python
linReg.score(train_features, train_typeCats)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(train_typeCats, linReg.predict(train_features))
```

```python
mean_squared_error(test_typeCats, linReg.predict(test_features))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decTreePartial1 = DecisionTreeClassifier(random_state=0, max_depth=3).fit(train_features, train_typeCats)
```

```python
from sklearn import tree
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
```

```python
graph1 = Source(tree.export_graphviz(decTreePartial1, out_file=None, feature_names=names, class_names=['1','2','3','4','5','6','7'], filled=True))
graph1a = Source(tree.export_graphviz(decTreePartial1, out_file='decTree1.dot', feature_names=names, class_names=['1','2','3','4','5','6','7'], filled=True))

display(SVG(graph1.pipe(format='svg')))
```

```python
decTreePartial2 = DecisionTreeClassifier(random_state=0, max_depth=4).fit(train_features, train_typeCats)
```

```python
decTreeFull = DecisionTreeClassifier(random_state=0).fit(train_features, train_typeCats)
graph = Source(tree.export_graphviz(decTreeFull, out_file='decTree.dot', feature_names=names, class_names=['1','2','3','4','5','6','7'], filled=True))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randFor = RandomForestClassifier(n_estimators=10000, n_jobs=6, random_state=43).fit(train_features, train_typeCats)
```

```python
importances = randFor.feature_importances_
indices2 = np.argsort(importances)[::-1]
```

```python
for f in range(features.shape[1]):
    print("%d. feature %d (%f) %s" % (f + 1, indices2[f], importances[indices2[f]], names[indices2[f]]))
```

```python
sortedImportances = np.flip(np.sort(importances))
```

```python
plt.plot(sortedImportances)
plt.savefig('randForImpt.png')
```

```python
randFor.score(test_features, test_typeCats)
```

```python
mean_squared_error(test_typeCats, randFor.predict(test_features))
```
