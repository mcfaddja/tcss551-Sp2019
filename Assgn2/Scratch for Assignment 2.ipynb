```python
xs = np.linspace(-np.pi, np.pi, 100)
ys = np.cos(xs)

print(len(xs))
```

```python
def derivative(xIN, yIN):
    xCNT = len(xIN)
    yCNT = len(yIN)
    
    if xCNT != yCNT:
        return 0
    else:
        cmpSLOPE = []
        cmpPOINT = []
        
        cnt = xCNT - 1
        for i in range(cnt):
            xTMP = xIN[i+1] - xIN[i]
            yTMP = yIN[i+1] - yIN[i]
            
            tmpSLOPE = yTMP / xTMP
            tmpPOINT = xIN[i] + xTMP / 2
            
            cmpSLOPE.append(tmpSLOPE)
            cmpPOINT.append(tmpPOINT)
            
        slopeOUT = np.asarray(cmpSLOPE)
        pointOUT = np.asarray(cmpPOINT)
        
        output = [slopeOUT, pointOUT]
        return output
```

```python
derivative(xs, ys)
```
